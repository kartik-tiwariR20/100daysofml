```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('/content/concrete_data.csv')
df.sample(5)
```

```python
df.isnull().sum()
```

```python
df.describe()
```

```python
from sklearn.model_selection import train_test_split,cross_val_score
x_train,x_test,y_train,y_test = train_test_split(df.drop('Strength',axis=1),df['Strength'],test_size=0.1,random_state=42)
y_train
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
lr = LinearRegression()
lr.fit(x_train,y_train)
y_pred = lr.predict(x_test)
r2_score(y_test,y_pred)
```

```python
np.mean(cross_val_score(lr,df.drop('Strength',axis=1),df['Strength'],scoring = 'r2'))
```

```python
from sklearn.preprocessing import PowerTransformer
ft = PowerTransformer(method='yeo-johnson')
x_train_transformed = ft.fit_transform(x_train)
x_test_transformed = ft.transform(x_test)
pd.DataFrame({'cols' : x_train.columns,'yeo-johnson_lambdas' : ft.lambdas_})
```

```python
lr.fit(x_train_transformed,y_train)
y_pred1 = lr.predict(x_test_transformed)
r2_score(y_test,y_pred1)
```

```python
np.mean(cross_val_score(lr,x_train_transformed,y_train,scoring = 'r2'))
```
