```python
import numpy as np
import pandas as pd
```

```python
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import OneHotEncoder
from sklearn.preprocessing import OrdinalEncoder
```

```python
df = pd.read_csv('/content/covid_toy.csv')
df.head()
```

```python
df.isnull().sum()
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(df.drop('has_covid',axis = 1),df['has_covid'],test_size = 0.1)
x_train.sample(5)
```

```python
from sklearn.compose import ColumnTransformer
transformer = ColumnTransformer(transformers = [
    ('tf1' , SimpleImputer(),['fever']),
    ('tf2' , OneHotEncoder(drop= 'first',sparse_output= False),['gender','city']),
    ('tf3' , OrdinalEncoder(categories = [['Mild','Strong']]),['cough'])
] , remainder='passthrough')
transformer.fit_transform(x_train)
```
