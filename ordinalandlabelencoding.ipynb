```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('/content/customer.csv')
df.sample(5)
```

```python
df = df.iloc[:,2:]
df.sample(5)
```

```python

from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(df.drop('purchased',axis=1),df['purchased'],test_size=0.2)
x_train.sample(5),x_test.sample(5)
```

```python
from sklearn.preprocessing import OrdinalEncoder
oe = OrdinalEncoder(categories=[['Poor','Average','Good'],['School','UG','PG']])
x_train = oe.fit_transform(x_train[['review','education']])
x_test = oe.transform(x_test[['review','education']])
```
