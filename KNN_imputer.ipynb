```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('/content/train.csv',usecols = ['Age','Fare','Survived','Pclass'])
df.sample(10)
```

```python
df.isnull().mean()*100
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(df.drop('Survived',axis=1),df['Survived'],test_size=0.1)
x_train
```

```python
from sklearn.impute import SimpleImputer,KNNImputer
from sklearn.compose import ColumnTransformer
knn = KNNImputer(weights='distance',n_neighbors=2)
si = SimpleImputer(strategy='mean')
trf = ColumnTransformer([
    ('age_imputed',si,['Age'])
],remainder='passthrough')
x_train1 = knn.fit_transform(x_train)
x_train2 = trf.fit_transform(x_train)
x_test1 = knn.transform(x_test)
x_test2 = trf.transform(x_test)
```

```python
from sklearn.linear_model import LogisticRegression
lr = LogisticRegression()
lr.fit(x_train1,y_train)
lr.fit(x_train2,y_train)

y_pred1 = lr.predict(x_test1)
y_pred2 = lr.predict(x_test2)
```

```python
from sklearn.metrics import accuracy_score
print(accuracy_score(y_test,y_pred1))
print(accuracy_score(y_test,y_pred2))
```
