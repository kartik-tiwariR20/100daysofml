```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
from sklearn.model_selection import train_test_split,cross_val_score
from sklearn.metrics import accuracy_score
from sklearn.compose import ColumnTransformer
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.preprocessing import FunctionTransformer
```

```python
df = pd.read_csv('/content/train.csv',usecols=['Age','Fare','Survived'])
df.sample(5)
```

```python
df.isnull().sum()
```

```python
df['Age'].fillna(df['Age'].mean(),inplace=True)
```

```python
df.isnull().sum()
```

```python
df['Age'].skew()
```

```python
df['Fare'].skew()
```

FARE IS RIGHT SKEWED AND AGE IS NORMALLY DISTRIBUTED


```python
x_train,x_test,y_train,y_test = train_test_split(df.drop('Survived',axis = 1),df['Survived'],test_size=0.1,random_state=42)
x_train
```

```python
clf1 = LogisticRegression()
clf2 = DecisionTreeClassifier()

clf1.fit(x_train,y_train)
clf2.fit(x_train,y_train)

y_pred1 = clf1.predict(x_test)
y_pred2 = clf2.predict(x_test)

print("LR",accuracy_score(y_test,y_pred1))
print("DT",accuracy_score(y_test,y_pred2))
```

```python
tf1 = ColumnTransformer([
    ('log',FunctionTransformer(np.log1p),['Fare'])
], remainder='passthrough')

x_train_transformed = tf1.fit_transform(x_train)
x_test_transformed = tf1.transform(x_test)
```

```python
clf1.fit(x_train_transformed,y_train)
clf2.fit(x_train_transformed,y_train)

y_pred_trans1 = clf1.predict(x_test_transformed)
y_pred_trans2 = clf2.predict(x_test_transformed)

print("LR",accuracy_score(y_test,y_pred_trans1))
print("DT",accuracy_score(y_test,y_pred_trans2))

```
