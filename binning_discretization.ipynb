```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('/content/train.csv',usecols=['Age','Fare','Survived'])
df.describe()
```

```python
df.dropna(inplace=True)
```

```python
df.isnull().sum()
```

```python
df.sample(5)
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(df.drop('Survived',axis=1),df['Survived'],test_size = 0.1)
x_train
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score
clf = DecisionTreeClassifier()
clf.fit(x_train,y_train)
y_pred = clf.predict(x_test)
accuracy_score(y_test,y_pred)
```

```python
from sklearn.preprocessing import KBinsDiscretizer
from sklearn.compose import ColumnTransformer
def discretization(bins,strategy):
  k_bin_age = KBinsDiscretizer(n_bins=bins,encode='ordinal',strategy=strategy)
  k_bin_fare = KBinsDiscretizer(n_bins=bins,encode='ordinal',strategy=strategy)

  tf = ColumnTransformer([
      ('first',k_bin_age,[0]),
      ('second',k_bin_fare,[1])
  ])
  x_train_trf=tf.fit_transform(x_train)
  x_test_trf=tf.transform(x_test)

  clf.fit(x_train_trf,y_train)
  y_pred1=clf.predict(x_test_trf)
  c = accuracy_score(y_test,y_pred1)
  print(c)
```

```python
discretization(10,'quantile')
```
